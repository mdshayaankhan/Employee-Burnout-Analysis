```python
nan_cols=data.columns[data.isnull().any()].tolist()
nan_cols_dtypes=data[nan_cols].dtypes
nan_cols_dtypes
```

```python
#############################filling the NaN column


data = data.apply(lambda col: col.fillna(col.mean()) if col.dtype in ['float64', 'int64'] else col)
```

```python
###using get_dummies to change string type into integer
if all (col in data.columns for col in ['Company Type','WFH Setup Available','Gender']):
  data=pd.get_dummies(data,columns=['Company Type','WFH Setup Available','Gender'])
else:
  print("columns not found")
```

```python
data.head(-1)
```

```python
##work id data Sciencetist

X=data.drop('Burn Rate',axis=1)
y=data['Burn Rate']
```

```python

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,shuffle=True,random_state=2)

#Scaling X independent

Scaler=StandardScaler()
Scaler.fit(X_train)
X_train=pd.DataFrame(Scaler.transform(X_train),index=X_train.index,columns=X_train.columns)
X_test=pd.DataFrame(Scaler.transform(X_test),index=X_test.index,columns=X_test.columns)
```

```python
X_train
```

```python
linear_regression_model=LinearRegression()
linear_regression_model.fit(X_train,y_train)
```

```python
###Linera regression model performance method

print('The Linear Regression model performance for training set')
y_predit=linear_regression_model.predict(X_test)

##Calculation of Mean square
mse=mean_squared_error(y_test,y_predit)
print('The Mean Square error is:',mse)

###Calculation of root mean square error
rmse=mean_squared_error(y_test,y_predit,squared=False)
print('The Root Mean Square is:',rmse)

###Calculation of mean absolute error
mae=mean_absolute_error(y_test,y_predit)
print('The Mean Absolute error is:',mae)

###Calculation of r2 score
r2=r2_score(y_test,y_predit)
print('The r2 score is:',r2)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import accuracy_score

knn=KNeighborsRegressor(n_neighbors=5)
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid to search
param_grid = {'n_neighbors': range(1, 21),  # Try k values from 1 to 20
              'metric': ['euclidean', 'manhattan']}

# Create a GridSearchCV object
grid_search = GridSearchCV(KNeighborsRegressor(), param_grid, cv=5, scoring='r2')

# Fit the grid search to your training data
grid_search.fit(X_train, y_train)

# Get the best model and its parameters
best_knn = grid_search.best_estimator_
print(f"Best parameters: {grid_search.best_params_}")

# Evaluate the best model
k_predict = best_knn.predict(X_test)
accuracy = r2_score(y_test, k_predict)
print(f"R-squared (after tuning): {accuracy}")
```
